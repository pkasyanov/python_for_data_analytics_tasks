1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Лінійна функція
f = 10 * x - 5
derivative_f = sp.diff(f, x)
print(derivative_f)
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = [10] * len(x) # Похідна f'(x) = 10

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Оцінка зростає з годинами. Кут нахилу $10$ показує, скільки балів у середньому дає одна година навчання, тобто кожна година вивчення підвищує оцінку на $10$ балів

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Визначаємо символічну змінну
x = sp.symbols('x')

f_2 = sp.sin(x)
derivative_f_2 = sp.diff(f_2, x)
print(derivative_f_2)
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = np.sin(x)
df = np.cos(x)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label="f(x) = sin(x)", color='blue')
plt.plot(x, df, label="f'(x) = cos(x)", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції f(x)=sin(x) та її похідної f'(x)=cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

f_3 = sp.ln(x)
derivative_f_3 = sp.diff(f_3, x)
print(derivative_f_3)
```

```python
# Визначення функції та її похідної
x = np.linspace(0.1, 10, 100)
f = np.log(x)
df = 1 / x

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label="$f(x) = ln(x)$", color='blue')
plt.plot(x, df, label="$f'(x)=\\frac{1}{x}$", color='green')

# Додавання легенди, міток та заголовка
plt.title(r"Графік функції $f(x)=\ln(x)$ та її похідної $f'(x)=\frac{1}{x}$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

f_4 = x ** 2 - 5 * x + 1
derivative_f_4 = sp.diff(f_4, x)
print(derivative_f_4)
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 100, 1000)
f = x ** 2 - 5 * x + 1
df = 2 * x - 5

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label="f(x) = $x^2 - 5x + 1$", color='blue')
plt.plot(x, df, label="$f'(x) = 2x - 5$", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції $f(x)=x^2 - 5x + 1$ та її похідної $f'(x)=2x-5$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
